```python
#import new package
import numpy as np
import pandas as pd
```

```python
#initalize a list of the 100 integers starting from 0 until 99
random_list = np.arange(100)
```

```python
#print list 
print(random_list)
```

```python
#acess elements within the array
print(random_list [0])
```

```python
#the : operator which can be used to acess elements across ranges
print(random_list [5:15])
```

```python
#use : at the start or end of an array to print out "everything from start until" and everything __ until the end
print(random_list[:5])
```

```python
print(random_list [5:])
```

```python
#We can also do math on our array!
random_list_x2 = random_list*2
print(random_list_x2)

random_list = random_list+1
print(random_list)
```

```python
random_list_log = np.log(random_list)
print(random_list_log)
```

```python
#boolean (true/false) index our array to print out all even numbers
print(random_list % 2 == 0)
even_list = random_list[random_list % 2 == 0]
print(even_list)
```
